# 04. Backtest Analysis

Load the JSON written by `scripts/run_backtest.py` and visualize key metrics.

```python
import json
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd

results = json.loads(Path('../data/backtest_results.json').read_text())
report = results['report']
report
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
for k, v in results['prediction_distribution'].items():
    ax.bar([k], [v])
ax.set_title('Prediction distribution on test set')
plt.show()
```

```python
metrics = ['sharpe', 'sortino', 'max_drawdown', 'hit_rate', 'profit_factor']
fig, ax = plt.subplots(figsize=(8, 4))
vals = [float(report[m]) for m in metrics]
ax.bar(metrics, vals)
ax.set_title('Backtest performance')
plt.xticks(rotation=20); plt.show()
```
