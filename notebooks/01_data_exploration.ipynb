# 01. Data Exploration

Load the synthetic / LOBSTER snapshots and explore the raw order book.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path('..').resolve()))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from src.data.synthetic import SyntheticLOBGenerator

gen = SyntheticLOBGenerator(n_events=20_000, seed=7)
ob, msg = gen.generate()
print(ob.shape, msg.shape)
ob.head()
```

```python
mid = (ob['ask_price_1'] + ob['bid_price_1']) / 2
spread = ob['ask_price_1'] - ob['bid_price_1']
fig, ax = plt.subplots(2, 1, figsize=(10, 6), sharex=True)
ax[0].plot(mid.values)
ax[0].set_title('Mid-price (synthetic)')
ax[1].plot(spread.values, color='orange')
ax[1].set_title('Spread (synthetic)')
plt.tight_layout(); plt.show()
```

```python
# Depth profile (mean size per level)
levels = range(1, 11)
bid_means = [ob[f'bid_size_{i}'].mean() for i in levels]
ask_means = [ob[f'ask_size_{i}'].mean() for i in levels]
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(list(levels), bid_means, marker='o', label='bid')
ax.plot(list(levels), ask_means, marker='s', label='ask')
ax.set_xlabel('Level'); ax.set_ylabel('Mean size'); ax.legend()
plt.show()
```
